```python
df = mvader_com
df
```

```python
date_unique = df['created_utc'].unique()
date_unique
```

```python
symbol_list = df.Symbol.tolist()
symbol_list
```

```python
# symbol_unique = df['Symbol'].unique()
symbol_unique = ['GME', 'AMC', 'BB', 'PLTR', 'TSLA', 'RH', 'NOK', 'RKT', 'AMD', 'TLRY']
# from collections import Counter
# c = Counter(symbol_list)
# symbol_unique = [key for key, value in c.most_common(500)]
symbol_unique
```

```python
empty_list = []
for i in date_unique:
    empty_list.append(0)
default_symbol_count = {}
for symbol in symbol_unique:
    default_symbol_count[symbol] = empty_list
default_symbol_count
```

```python
output_df = pd.DataFrame.from_dict(date_unique)
output_df.columns = ['created_utc']
output_df = output_df.append(default_symbol_count, ignore_index=True)
output_df = output_df.dropna(subset=['created_utc'])
```

```python
%%time
for date in date_unique:
    print(f"{date} is being processed")
    for symbol in symbol_unique:
        try:
            temp10 = mvader_com[(mvader_com['created_utc'] == date) & (mvader_com['Symbol'] == symbol)]
            output_mvader_com.loc[(output_mvader_com['created_utc'] == date), symbol] = temp10['Modified_Vader_Compound'].mean()
        except Exception as e:
            print("mvader_com", date, symbol, e)
```
