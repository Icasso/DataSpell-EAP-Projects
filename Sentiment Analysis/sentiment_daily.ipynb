```python
import pandas as pd

tbplarity = pd.read_pickle('../Pickles/symbol_textblob_polarity.pkl')
tbsubjectivity = pd.read_pickle('../Pickles/symbol_textblob_subjectivity.pkl')
dvader_neg = pd.read_pickle('../Pickles/symbol_dvader_negative.pkl')
dvader_pos = pd.read_pickle('../Pickles/symbol_dvader_positive.pkl')
dvader_neu = pd.read_pickle('../Pickles/symbol_dvader_neutral.pkl')
dvader_com = pd.read_pickle('../Pickles/symbol_dvader_compound.pkl')
mvader_neg = pd.read_pickle('../Pickles/symbol_mvader_negative.pkl')
mvader_pos = pd.read_pickle('../Pickles/symbol_mvader_positive.pkl')
mvader_neu = pd.read_pickle('../Pickles/symbol_mvader_neutral.pkl')
mvader_com = pd.read_pickle('../Pickles/symbol_mvader_compound.pkl')
```

```python
df = tbplarity
df
```

```python
date_unique = df['created_utc'].unique()
date_unique
```

```python
symbol_list = df.Symbol.tolist()
symbol_list
```

```python
from collections import Counter
c = Counter(symbol_list)
c.most_common(50)
```

```python
# symbol_unique = df['Symbol'].unique()
# symbol_unique = ['GME', 'AMC', 'BB', 'PLTR', 'TSLA', 'RH', 'NOK', 'RKT', 'AMD', 'TLRY']
symbol_unique = [key for key, value in c.most_common(50)]
symbol_unique
```

```python
output_df = pd.read_pickle('../Pickles/comment_volume_df.pkl')
output_df = output_df.drop(columns=['comment_volume'])
output_df
```

```python
output_tbplarity = output_df
output_tbsubjectivity = output_df
output_dvader_neg = output_df
output_dvader_pos = output_df
output_dvader_neu = output_df
output_dvader_com = output_df
output_mvader_neg = output_df
output_mvader_pos = output_df
output_mvader_neu = output_df
output_mvader_com = output_df
```

```python
%%time
for date in date_unique:
    print(f"{date} is being processed")
    for symbol in symbol_unique:
        try:
            temp = tbplarity[(tbplarity['created_utc'] == date) & (tbplarity['Symbol'] == symbol)]
            output_tbplarity.loc[(output_tbplarity['created_utc'] == date), symbol] = temp['TextBlob_Polarity'].mean()
        except Exception as e:
            print("tbplarity", date, symbol, e)
        try:
            temp2 = tbsubjectivity[(tbsubjectivity['created_utc'] == date) & (tbsubjectivity['Symbol'] == symbol)]
            output_tbsubjectivity.loc[(output_tbsubjectivity['created_utc'] == date), symbol] = temp2['TextBlob_Subjectivity'].mean()
        except Exception as e:
            print("tbsubjectivity", date, symbol, e)
        try:
            temp3 = dvader_neg[(dvader_neg['created_utc'] == date) & (dvader_neg['Symbol'] == symbol)]
            output_dvader_neg.loc[(output_dvader_neg['created_utc'] == date), symbol] = temp3['Default_Vader_Negative'].mean()
        except Exception as e:
            print("dvader_neg", date, symbol, e)
        try:
            temp4 = dvader_pos[(dvader_pos['created_utc'] == date) & (dvader_pos['Symbol'] == symbol)]
            output_dvader_pos.loc[(output_dvader_pos['created_utc'] == date), symbol] = temp4['Default_Vader_Positive'].mean()
        except Exception as e:
            print("dvader_pos", date, symbol, e)
        try:
            temp5 = dvader_neu[(dvader_neu['created_utc'] == date) & (dvader_neu['Symbol'] == symbol)]
            output_dvader_neu.loc[(output_dvader_neu['created_utc'] == date), symbol] = temp5['Default_Vader_Neutral'].mean()
        except Exception as e:
            print("dvader_neu", date, symbol, e)
        try:
            temp6 = dvader_com[(dvader_com['created_utc'] == date) & (dvader_com['Symbol'] == symbol)]
            output_dvader_com.loc[(output_dvader_com['created_utc'] == date), symbol] = temp6['Default_Vader_Compound'].mean()
        except Exception as e:
            print("dvader_com", date, symbol, e)
        try:
            temp7 = mvader_neg[(mvader_neg['created_utc'] == date) & (mvader_neg['Symbol'] == symbol)]
            output_mvader_neg.loc[(output_mvader_neg['created_utc'] == date), symbol] = temp7['Modified_Vader_Negative'].mean()
        except Exception as e:
            print("mvader_neg", date, symbol, e)
        try:
            temp8 = mvader_pos[(mvader_pos['created_utc'] == date) & (mvader_pos['Symbol'] == symbol)]
            output_mvader_pos.loc[(output_mvader_pos['created_utc'] == date), symbol] = temp8['Modified_Vader_Positive'].mean()
        except Exception as e:
            print("mvader_pos", date, symbol, e)
        try:
            temp9 = mvader_neu[(mvader_neu['created_utc'] == date) & (mvader_neu['Symbol'] == symbol)]
            output_mvader_neu.loc[(output_mvader_neu['created_utc'] == date), symbol] = temp9['Modified_Vader_Neutral'].mean()
        except Exception as e:
            print("mvader_neu", date, symbol, e)
        try:
            temp10 = mvader_com[(mvader_com['created_utc'] == date) & (mvader_com['Symbol'] == symbol)]
            output_mvader_com.loc[(output_mvader_com['created_utc'] == date), symbol] = temp10['Modified_Vader_Compound'].mean()
        except Exception as e:
            print("mvader_com", date, symbol, e)
    print(f"{date} is done")
```

```python
output_tbplarity.to_pickle('../Pickles/out_symbol_textblob_polarity.pkl')
output_tbsubjectivity.to_pickle('../Pickles/out_symbol_textblob_subjectivity.pkl')
output_dvader_neg.to_pickle('../Pickles/out_symbol_dvader_negative.pkl')
output_dvader_pos.to_pickle('../Pickles/out_symbol_dvader_positive.pkl')
output_dvader_neu.to_pickle('../Pickles/out_symbol_dvader_neutral.pkl')
output_dvader_com.to_pickle('../Pickles/out_symbol_dvader_compound.pkl')
output_mvader_neg.to_pickle('../Pickles/out_symbol_mvader_negative.pkl')
output_mvader_pos.to_pickle('../Pickles/out_symbol_mvader_positive.pkl')
output_mvader_neu.to_pickle('../Pickles/out_symbol_mvader_neutral.pkl')
output_mvader_com.to_pickle('../Pickles/out_symbol_mvader_compound.pkl')
```
