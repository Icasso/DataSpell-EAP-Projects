```python
from textblob import TextBlob
import pandas as pd
```

```python
comments_df = pd.read_pickle("../Playground/comments_with_symbol_df.pkl")
```

```python
body_list = comments_df.body.tolist()
```

```python
textblob_polarity = []
textblob_subjectivity = []
for s in body_list:
    textblob_polarity.append(TextBlob(s).sentiment.polarity)
    textblob_subjectivity.append(TextBlob(s).sentiment.subjectivity)
```

```python
comments_df["TextBlob_Polarity"] = textblob_polarity
comments_df["TextBlob_Subjectivity"] = textblob_subjectivity
comments_df
```
