```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from data import *
```

```python
comments_df = pd.read_pickle("comments_with_textblob_dvader_df.pkl")
# comments_df = comments_df[0:1000]
```

```python
comments_df
```

```python
body_list = comments_df.body.tolist()
```

```python
vader = SentimentIntensityAnalyzer()
vader.lexicon.update(new_words)
```

```python
rsi_vader = []
```

```python
%%time
for s in body_list:
    rsi_vader.append(vader.polarity_scores(s))
```

```python
rsi_vader
```

```python
len(rsi_vader)
```

```python
rsi_neg = []
rsi_neu = []
rsi_pos = []
rsi_compound = []
for i in rsi_vader:
    rsi_neg.append(i['neg'])
    rsi_neu.append(i['neu'])
    rsi_pos.append(i['pos'])
    rsi_compound.append(i['compound'])
```

```python
comments_df["Modified_Vader_Negative"] = rsi_neg
comments_df["Modified_Vader_Neutral"] = rsi_neu
comments_df["Modified_Vader_Positive"] = rsi_pos
comments_df["Modified_Vader_Compound"] = rsi_compound
comments_df
```

```python
comments_df.to_pickle('comments_fin_df.pkl')
```
