```python
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
comments_df = pd.read_pickle("comments_with_textblob_df.pkl")
# comments_df = comments_df[0:1000]
```

```python
comments_df
```

```python
body_list = comments_df.body.tolist()
```

```python
vader = SentimentIntensityAnalyzer()
```

```python
default_vader = []
```

```python
%%time
for s in body_list:
    default_vader.append(vader.polarity_scores(s))
```

```python
default_vader
```

```python
len(default_vader)
```

```python
neg = []
neu = []
pos = []
compound = []
for i in default_vader:
    neg.append(i['neg'])
    neu.append(i['neu'])
    pos.append(i['pos'])
    compound.append(i['compound'])
```

```python
comments_df["Default_Vader_Negative"] = neg
comments_df["Default_Vader_Neutral"] = neu
comments_df["Default_Vader_Positive"] = pos
comments_df["Default_Vader_Compound"] = compound
comments_df
```

```python
comments_df.to_pickle('comments_with_textblob_dvader_df.pkl')
```
